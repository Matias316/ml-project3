```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
corona_df = pd.read_csv('datasets/Corona_with_temperatures.csv')
corona_df.head()
```

```python
corona_df['weather'] = np.where(corona_df['avg_temp'] >= 25, 1,0)
print(corona_df.head())
```

```python
confirmed_hot_climate_df = corona_df[corona_df['weather'] == 1]
confirmed_cold_climate_df = corona_df[corona_df['weather'] == 0]
```

```python
var_hot_climate = np.var(confirmed_hot_climate_df['avg_temp'])
var_cold_climate = np.var(confirmed_cold_climate_df['avg_temp'])
print(var_hot_climate, var_cold_climate)
print("Ratio:" + str(var_cold_climate/var_hot_climate))
```

```python
#Because Ratio is greater than 4:1 we will use Welch's t-test
stats.ttest_ind(a=confirmed_hot_climate_df['avg_temp'], b=confirmed_cold_climate_df['avg_temp'],equal_var=False)
```
